VISUALIZANDO A MÚSICA COM MAIOR DURAÇÃO.

```python
# Encontrar o índice da música com o maior duração
indice_maior_duração = data_spotfy['duration_ms'].idxmax()

# Usar o índice para obter o nome da música com o maior duração
musica_maior_duração = data_spotfy.loc[indice_maior_duração, 'track_name']

# Exibir o nome da música com o maior duração
print(f'A música com o maior duração é: {musica_maior_duração} com {indice_maior_duração} milisegundos')
```

SELECIONANDO AS 100 MÚSICAS MAIS POPULARES E EXIBINDO APENAS O TOP 5

```python
top_100_musicas = data_spotfy.nlargest(100, 'popularity')
top_100_musicas.head(5)
```

DAY 02


```python
# Definindo a condição
condicoes = [data_spotfy['popularity'] > 70]
valores = [1]

# Adicionando a condição de não popular
condicoes.append(data_spotfy['popularity'] <= 70)
valores.append(0)

# Criando a nova coluna 'most_popularity'
data_spotfy['most_popularity'] = np.select(condicoes, valores, default=0)

# Exibindo o resultado
data_spotfy[['popularity', 'most_popularity']].head(15)
```
